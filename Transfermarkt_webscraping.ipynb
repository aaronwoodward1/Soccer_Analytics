We set a headers variable for us to get permission to access the Transfermarkt website. Then, we set the url to the position pages that we're trying to acquire data from. Once we have all the elements listed in the bulletpoints mentioned above, we construct a pandas DataFrame with those elements. 

```python
playerids = []
names = []
values = []
nats = []
clubs = []
position = []
ages = []

headers= {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

url = url_cf #page you need to change position abbrevs to get data from different positions.
    
r = requests.get(url, headers=headers) #pageTree

soup = BeautifulSoup(r.text, 'html.parser')
pretty = soup.prettify()

for table in soup.find_all('table', attrs={'class': 'inline-table'}):
    content = table.contents
    names.append(content[0].text)  # Name
    position.append(content[1].text)  # Position

#for td in soup.find_all('td', attrs={'clacb': 'zentriert'}):
#    agetext = td.text
#    ages.append(agetext[4].text)
    
#for i in soup.findAll("td", clacb_="zentriert")[1]: #Age
#    ages.append(i.text)
    
#for td in soup.findAll("td", clacb_="zentriert"): #Age
#    ages.append(content[1].text)

for value in soup.findAll("td", class_="rechts hauptlink"): #Value
    values.append(value.get_text(strip=True))
    
for td in soup.findAll("td", class_="zentriert"): #Nationality
    inner_grp = []
    for item in td.findAll("img", class_="flaggenrahmen", title=re.compile("^(?!http).*")):
        if item.get('title'):
            inner_grp.append(item.get('title'))
    if inner_grp:
        nats.append(inner_grp)
    
for club in soup.findAll("img", src=re.compile("^http"), class_="", title=True)[:25]: #Club
    clubs.append(club.get("alt"))

#for i in range(0,25): #Age
#    ages.append(soup.findAll(clacb_='zentriert')[2].text) 
    
    
#for i in range(0,25): #Age
#    ages.append(soup.findAll('td')[0].text)    
    
#for item in soup.findAll("a", clacb_="spielprofil_tooltip tooltipstered")[:25]:
#    playerids.append(item.get_text("id"))
    
tm_df_cf1 = pd.DataFrame({"Player":names, "Club":clubs, "Nationality":nats,"Position":position,"Value":values})

tm_df_cf1.head(25)
```

Due to the behavior of the Transfermarkt website, its very difficult to scrape pages all at once, beginning with first page. However, after the first page, the behavior of the website is more stable. Therefore, we can iterate through page 2 and 21 (final page of the position url). We'll make a few tweaks to the code above to incorporate the iteration of scraping the remaining pages.

```python
#Source: https://stackoverflow.com/questions/60780519/how-do-i-scrape-multiply-country-images-from-players-on-transfermarkt-and-still
headers= {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}
    

pages = list(range(2, 21))

playerids = []
names = []
values = []
nats = []
clubs = []
position = []



for page in pages:
    url = url_lb2 + str(page)
    
    r = requests.get(url, headers=headers)

    soup = BeautifulSoup(r.text, 'html.parser')
    pretty = soup.prettify()

    for table in soup.find_all('table', attrs={'class': 'inline-table'}):
        content = table.contents
        names.append(content[0].text)  # Name
        position.append(content[1].text)  # Position
        
    for value in soup.findAll("td", class_="rechts hauptlink"): #Value
        values.append(value.get_text(strip=True))

    for td in soup.findAll("td", class_="zentriert"): #Nationality
        inner_grp = []
        for item in td.findAll("img", class_="flaggenrahmen", title=re.compile("^(?!http).*")):
            if item.get('title'):
                inner_grp.append(item.get('title'))
        if inner_grp:
            nats.append(inner_grp)

    for club in soup.findAll("img", src=re.compile("^http"), class_="", title=True)[:25]: #Club
        clubs.append(club.get("alt"))
        
    for td in soup.findAll("a", class_="spielprofil_tooltip tooltipstered")[:25]:
        playerids.append(td.get("id"))
        
tm_df_lb2 = pd.DataFrame({"Player":names, "Club":clubs, "Nationality":nats,"Position":position,"Value":values})

tm_df_lb2.head(50)
```

### Data Transformation for Value column

In order to use the player market value for analysis, the "Value" column needs to undergo a data transformation. The column is recognized as a string or character because of the currency symbols (€, $, £, etc.) and value units suffix (m, Th). There are number of steps that we take to do this.

### First, we add a column "Value_suffix"

```python
#Source: https://www.pythonprogramming.in/if-value-in-row-in-dataframe-contains-string-create-another-column-equal-to-string-in-pandas.html
tm_df_master['Value_suffix']= pd.np.where(tm_df_master.Value.str.contains("m"), "m", "k")
tm_df_master.head(500)                                          
```

### Second, we replace the currency symbols and value suffixes from Value column

```python
tm_df_master['Value'] = tm_df_master['Value'].str.replace('€', '')
tm_df_master['Value'] = tm_df_master['Value'].str.replace('m', '')
#tm_df_master['Nationality'] = tm_df_master['Nationality'].str.replace('[', '')
tm_df_master['Value'] = tm_df_master['Value'].str.replace('k', '')
#tm_df_master['Value'] = tm_df_master['Value'].astype(float)
```

```python
tm_df_master.head(500)
```

### Converting Value to float datatype

```python
tm_df_master['Value'] = tm_df_master['Value'].astype(float)
```

### Using the Value_suffix column to evaluate which values that we have to convert to thousands of euros.

```python
tm_df_master['Value_EU_M']= pd.np.where(tm_df_master.Value_suffix.str.contains("k"), tm_df_master['Value']/1000, tm_df_master['Value'])
tm_df_master.head(500)
```

### Subset the master dataframe by keeping the relevant columns

```python
#tm_df_master = tm_df_master.loc[:, ~tm_df_master.columns.str.contains(['Value', 'Value_suffix'])]
#tm_df_master = tm_df_master.drop(['Value','Value_suffix'], axis=1, inplace=True)

tm_df_master = tm_df_master[['Player', 'Club', 'Nationality', 'Position', 'Value_EU_M']]
tm_df_master.head()
```

Let's sort the master dataframe by player market value ("Value_EU_M") in descending order so that the highest value player appears first.

```python
tm_df_master = tm_df_master.sort_values('Value_EU_M', ascending=False)
tm_df_master.head(50)
```
